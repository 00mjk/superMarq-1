```python
import glob
import pandas as pd
import supermarq as sm
```

# GHZ

```python
title = 'GHZ'
labels = ['3 qubits', '7 qubits', '11 qubits']
feature_vecs = []
for nq in [3, 7, 11]:
    circ = sm.ghz.GHZ(nq).circuit()
    con = sm.features.compute_connectivity(circ)
    liv = sm.features.compute_liveness(circ)
    par = sm.features.compute_parallelism(circ)
    mea = sm.features.compute_measurement(circ)
    ent = sm.features.compute_entanglement(circ)
    dep = sm.features.compute_depth(circ)
    feature_vecs.append([con, liv, par, mea, ent, dep])
    
#spoke_labels = None
savefn = None
spoke_labels = ['Com', 'Liv', 'Par', 'Mea', 'Ent', 'Dep']
#savefn = 'figures/feature_maps/mermin-bell.pdf'
sm.features.plot_benchmark([title, labels, feature_vecs], savefn=savefn, spoke_labels=spoke_labels, legend_loc=(0.1, 0.25))
```

# Create a DataFrame with the application features

```python
all_pickles = glob.glob('benchmark_results/*pickle')
print(len(all_pickles))
benchmarks = sorted(list(set(['_'.join(name.split('/')[1].split('_')[0:2]) for name in all_pickles])))
print('Benchmarks:', benchmarks)
features = ['Communication', 'Liveness', 'Parallelism', 'Measurement', 'Entanglement', 'Depth', 'qubits', 'entangling-gates', 'regular-depth']
df = pd.DataFrame(data=None, index=benchmarks, columns=features)
df.head()
```

```python
# Fill in the dataframe
for application in df.index:
    name = application.split('_')[0]
    if name == 'ghz':
        nq = int(application.split('_')[1].strip('qubits'))
        bmark = sm.ghz.GHZ(nq)
    elif name == 'merminbell':
        nq = int(application.split('_')[1].strip('qubits'))
        bmark = sm.mermin_bell.MerminBell(nq)
    elif name == 'hamsim':
        nq = int(application.split('_')[1].split('q')[0])
        steps = int(application.split('_')[1].split('s')[1])
        bmark = sm.hamiltonian_simulation.HamiltonianSimulation(nq, 1, steps)
    elif name == 'qaoafswap':
        nq = int(application.split('_')[1].strip('qubits'))
        bmark = sm.qaoa_fermionic_swap_proxy.QAOAFermionicSwapProxy(nq)
    elif name == 'qaoavanilla':
        nq = int(application.split('_')[1].strip('qubits'))
        bmark = sm.qaoa_vanilla_proxy.QAOAVanillaProxy(nq)
    elif name == 'vqe':
        nq = int(application.split('_')[1].split('q')[0])
        layers = int(application.split('_')[1].split('s')[1].strip('layer'))
        bmark = sm.vqe_proxy.VQEProxy(nq, layers)
    elif name == 'bitcode':
        nq = int(application.split('_')[1].split('q')[0])
        rounds = int(application.split('_')[1].split('s')[1].strip('round'))
        bmark = sm.bit_code.BitCode(nq, rounds, [1]*nq)
    elif name == 'phasecode':
        nq = int(application.split('_')[1].split('q')[0])
        rounds = int(application.split('_')[1].split('s')[1].strip('round'))
        bmark = sm.phase_code.PhaseCode(nq, rounds, [1]*nq)

    for feature in df.columns:
        if name == 'vqe':
            circ = bmark.circuit()[0]
        else:
            circ = bmark.circuit()
        
        if feature == 'Communication':
            val = sm.features.compute_connectivity(circ)
        elif feature == 'Liveness':
            val = sm.features.compute_liveness(circ)
        elif feature == 'Parallelism':
            val = sm.features.compute_parallelism(circ)
        elif feature == 'Measurement':
            val = sm.features.compute_measurement(circ)
        elif feature == 'Entanglement':
            val = sm.features.compute_entanglement(circ)
        elif feature == 'Depth':
            val = sm.features.compute_depth(circ)
        elif feature == 'qubits':
            val = len(circ.all_qubits())
        elif feature == 'entangling-gates':
            qiskit_circ = sm.features.cirq_to_qiskit(circ)
            val = 0
            gate_counts = qiskit_circ.count_ops()
            val += gate_counts.get('cx', 0)
            val += gate_counts.get('cz', 0)
        elif feature == 'regular-depth':
            qiskit_circ = sm.features.cirq_to_qiskit(circ)
            val = qiskit_circ.depth()
        df.at[application, feature] = val
```
